```python
#Boxplot to view outliers in our dataset
num_variables = ["TOTAL NUMBER OF HIV DIAGNOSES","HIV DIAGNOSES PER 100,000 POPULATION","TOTAL NUMBER OF CONCURRENT HIV/AIDS DIAGNOSES","PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES","TOTAL NUMBER OF AIDS DIAGNOSES", "AIDS DIAGNOSES PER 100,000 POPULATION"]
def getBoxPlotforNumVar():
    try:
        val=int(input("enter the index of the numerical variable you intend to view its box plot:"))
        plt.figure(figsize=(10,10))
        sb.boxplot(df[num_variables[val]], data=df)
    except BaseException as error:
        print(error)
getBoxPlotforNumVar()
```

```python
#Confirming the positive skewness of the distribution
P_Mean = df['PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES'].mean()
P_Median = df['PROPORTION OF CONCURRENT HIV/AIDS DIAGNOSES AMONG ALL HIV DIAGNOSES'].median()
print("Mean Value is {} ".format(P_Mean))
print("Median Value is {}".format(P_Median))
```

```python
# method to replace missing values in all numerical variables in our dataset
def fillNaNValuesforNum_Var():
    try:
        for col_var in range(len(num_variables)):
            df[num_variables[col_var]].fillna(df[num_variables[col_var]].median(), inplace=True)
    except BaseException as error:
        print(error)
fillNaNValuesforNum_Var()
```

```python
#GET Category Plot of Genders Diagnosed with HIV and AIDS for 2019
bronx=df.loc[df['Borough'] == 'bronx'][['YEAR','Neighborhood (U.H.F)','SEX','TOTAL NUMBER OF HIV DIAGNOSES','TOTAL NUMBER OF AIDS DIAGNOSES']]
data_br = bronx.loc[(bronx['Neighborhood (U.H.F)'] == 'huntspoint-motthaven') & (bronx['YEAR'] == '2019-01-01')]
reindexed_data = data_br.groupby('SEX',as_index = False).sum()
reindexed_data.rename(columns={'SEX':'Sex','TOTAL NUMBER OF HIV DIAGNOSES':'Total-HIV-Dia','TOTAL NUMBER OF AIDS DIAGNOSES':'Total-AIDS-Dia'}, inplace=True)
def getBronx2019():
    plt.figure(figsize=(10,10))
    try:
        g = sb.catplot(
        data=reindexed_data.melt(id_vars = ["Sex"], value_vars=["Total-HIV-Dia", "Total-AIDS-Dia"]),
        kind="bar",
        y="value", x="variable", col="Sex", palette='Pastel2'
        )
        g.set_axis_labels("Infection Count (HIV & AIDS)", "Infection Type(HIV & AIDS")

    except BaseException as error:
        print(error)
getBronx2019()
```
